#DISTPLOT

A distplot plots a univariate distribution of observations.

```python
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(df['Price'])
plt.show()
```

```python
sns.distplot(df['KM'], kde = False, color ='red', bins =5)
```

#HISTPLOT

###VISUALISE THE UNIQUE COUNT

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

```python
sns.histplot(df['Price'],palette='hls')
plt.show()
```

#SCATTERPLOT

Scatter plots are the graphs that present the relationship between two variables in a data-set.

```python
import seaborn as sns
sns.scatterplot(x='Airbag_1',y='Airbag_2',data=df)
```

```python
sns.scatterplot(x='Price',y='KM',data=df)
```

#PAIRPLOT

A pair plot is a data visualization that plots pair-wise relationships between all the variables of a dataset.

```python
import seaborn as sns
sns.pairplot(df)
```

#COUNTPLOT

A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
sns.countplot(x='Color',data=df)
plt.show()
```

```python
sns.countplot(x='Fuel_Type',data=df)
plt.show()
```

#HEATMAP

A heatmap is a graphical representation of data that uses a system of color coding to represent different values.

```python
import seaborn as sns
sns.heatmap(df.corr(numeric_only="silence"),cmap='Greens',annot=True)
```

```python
data=np.random.rand(10,10)
plt.imshow(data,cmap='hot')
plt.show()
```

#LINEPLOT

A line plot is a type of plot in Seaborn that shows the relationship between two variables by connecting the data points with a straight line.

```python
sns.lineplot(x="Fuel_Type", y="HP", data=df)
plt.show()
```

```python
sns.lineplot(x="Price", y="HP", data=df)
plt.show()
```

#Violinplot

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data.

```python
sns.violinplot(x ="Fuel_Type", y ="Price", data =df,color='cyan')
```
